# 0. Setup Path

```python
import os
import shutil
```

```python
drive = 'H://Machine Learning//TFODCourse//'
s = '//'
```

```python
collectedimages = drive + s.join(['Tensorflow', 'workspace','images','collectedimages'])
collectedimages 
```

```python
img_path =  s.join([collectedimages , 'image'])
img_path
```

# 1. Select the image with config to transfer

```python
os.chdir(img_path)
```

```python
%ls
```

```python
all_img = []

for f in os.listdir("."):
    all_img.append(f)

all_img
```

```python
JPG =[]
XML =[]

for img in all_img:
    if ".xml" in img:
        XML.append(img)
        img = img.replace('xml','jpg')
        JPG.append(img)
```

```python
JPG
```

```python
XML
```

```python
train_test_p= 75/100
train_test_n = int(len(JPG)*train_test_p)
print(train_test_n, 'to be trained out of', len(JPG))
```

```python
JPG_train = JPG[:train_test_n]
JPG_test = JPG[train_test_n : ]

XML_train = XML[:train_test_n]
XML_test = XML[train_test_n: ]
```

```python
len(JPG_test)
```

# 2. Copy Selected files to Train & Test Files

```python
train = drive + s.join(['Tensorflow', 'workspace','images','train'])
train
```

```python
test = drive + s.join(['Tensorflow', 'workspace','images','test'])
test
```

```python
# Copy Train
for jpg, xml in zip(JPG_train, XML_train):
    dst_path = train
    
    src_path = s.join([img_path, jpg])
    shutil.copy(src_path, dst_path)
    print(jpg, 'Copied')
    
    src_path = s.join([img_path, xml])
    shutil.copy(src_path, dst_path)
    print(xml, 'Copied')
```

```python
# Copy Test
for jpg, xml in zip(JPG_test ,XML_test):
    dst_path = test
    
    src_path = s.join([img_path, jpg])
    shutil.copy(src_path, dst_path)
    print(jpg, 'Copied')
    
    src_path = s.join([img_path, xml])
    shutil.copy(src_path, dst_path)
    print(xml, 'Copied')
```
